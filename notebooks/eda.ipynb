# Population Stats

```python
summary_2000 = df['total_population_00'].describe()
summary_2017 = df['total_population_17'].describe()
```

```python
# Best to show max pop as all other data creates confusion
summary_df = pd.DataFrame({
        'Year': ['2000', '2017'],
        #'Mean': [summary_2000['mean'], summary_2017['mean']],
        #'Median': [summary_2000['50%'], summary_2017['50%']],
        #'Min': [summary_2000['min'], summary_2017['min']],
        'Max': [summary_2000['max'], summary_2017['max']],
        #'Std Dev': [summary_2000['std'], summary_2017['std']]
})
```

```python
fig = go.Figure()

for col in summary_df.columns[1:]:
        fig.add_trace(go.Bar(
                x=['2000'],
                y=[summary_df.loc[0,col]],
                name='2000',
                marker_color='rgb(142, 142, 142)'
                
        ))
        
        
        

for col in summary_df.columns[1:]:
        fig.add_trace(go.Bar(
                x=['2017'],
                y=[summary_df.loc[1,col]],
                name='2017',
                marker_color='rgb(205, 92, 92)'
                
        ))
        
        

fig.update_layout(
        title="<b> Summary Stats </b><br><sup> Total Population<sup>",
        xaxis=dict(title='Year'),
        yaxis=dict(title='Population'),
        barmode='group',
        plot_bgcolor='rgba(0, 0, 0, 0)'
)

fig.update_xaxes(tickfont=dict(size=12), title_text='', showticklabels=False)

fig.show()
```

# Median Income/Value

```python
median_income_2000 = df['median_income_00']
median_income_2017 = df['median_income_17']
```

```python
median_df = pd.DataFrame({
        'Year': ['2000', '2017'],
        'Median Income': [median_income_2000.median(), median_income_2017.median()]
        
})
```

```python
fig = go.Figure()

fig.add_traces(
        go.Bar(
                x=['2000'],
                y=[median_df.loc[0,'Median Income']],
                name='2000',
                marker_color='rgb(142, 142, 142)'
                
        )
)



fig.add_traces(
        go.Bar(
                x=['2017'],
                y=[median_df.loc[1,'Median Income']],
                name='2017',
                marker_color='rgb(205, 92, 92)'
                
        )
)

fig.update_layout(
        title='<b> Median Income Comparison </b><br><sup> 2000 vs 2017<sup>',
        xaxis=dict(title='Year'),
        yaxis=dict(title=''),
        barmode='group',
        plot_bgcolor='rgba(0, 0, 0, 0)'
)

fig.update_xaxes(tickfont=dict(size=12), title_text='', showticklabels=False)

fig.show()


```

# Educational Attainment

```python

educational_attainment_00 = df['educational_attainment_00']
educational_attainment_17 = df['educational_attainment_17']
```

```python
fig = go.Figure()

fig.add_traces(go.Bar(
        x=['2000'],
        y=[total_ed_df.loc[0,'Ed_Att']],
        name= '2000',
        marker_color='rgb(142, 142, 142)'
))

fig.add_traces(go.Bar(
        x=['2017'],
        y=[total_ed_df.loc[1,'Ed_Att']],
        name= '2000',
        marker_color='rgb(205, 92, 92)'
))

fig.update_layout(
        title="<b> Educational Attainment</b><br><b> 2000 vs 2017</b><br><sup> (25 or older with 4-year college degree)</sup>",
        xaxis=dict(title='Education'),
        yaxis=dict(title=''),
        barmode='group',
        plot_bgcolor='rgba(0, 0, 0, 0)'
)

fig.update_xaxes(tickfont=dict(size=12), title_text='', showticklabels=False)
```
